# Examples of the three presented types of bifurcation

Work based on Chapter 2 of the book [Nonlinear Dynamics and Chaos](http://www.stevenstrogatz.com/books/nonlinear-dynamics-and-chaos-with-applications-to-physics-biology-chemistry-and-engineering).

---------------------

* Saddle Node
 - $\hat{x} = r + x^2$
 
 
* Transcritical
 - $\hat{x} = r \cdot x - x^2$
 
 
* Pitchfork
 - $\hat{x} = r \cdot x - x^3$
 


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from collections import OrderedDict
```

### Function definitions of the three 'normal' types:

```python
def saddle(x, r): return r - x**2
def transcritical(x, r): return r * x - x**2
def pitchfork(x, r): return r * x - x**3
```

```python
# sample data range for plotting
x = np.arange(-2, 2, 0.1)
rs = [-1, 0, 1]
```

```python
# compute the values of the tiem derivatives
y_saddle = [saddle(x, r) for r in rs]
y_transritical = [transcritical(x, r) for r in rs]
y_pitchfork = [pitchfork(x, r) for r in rs]
```

```python
methods = ['saddle', 'transcritical', 'pitchfork']
colours = ['r', 'g', 'b']
leg_labs = ['r = {:>3}'.format(r) for r in rs]

fig, axs = plt.subplots(3)
fig.set_size_inches(8, 15)

for i, (method, x_hat) in enumerate(zip(methods, [y_saddle, y_transritical, y_pitchfork])):

    for j, v in enumerate(x_hat):
        axs[i].plot(x, v, c=colours[j])
    
    axs[i].set_xlim([-1.5, 1.5])
    axs[i].set_ylim([-1.5, 1.5])
    axs[i].set_title(method, fontsize=14)
    axs[i].grid(True)
    axs[i].legend(leg_labs)

```
